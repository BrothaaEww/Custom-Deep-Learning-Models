```python
import tensorflow as tf
import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Dense
import numpy
from keras.utils import plot_model
```

```python
model = keras.Sequential()
model.add(keras.Input(shape=(4,)))
model.add(layers.Dense(2, activation = 'relu'))
print(model.summary())
```

```python
plot_model(model, show_shapes = True, show_layer_names = True)
```

2 nd model

```python
model2 = keras.Sequential()
model2.add(keras.Input(shape=(4,)))
model2.add(layers.Dense(2, activation = 'relu'))
model2.add(layers.Dense(3, activation = 'relu'))
model2.add(layers.Dense(4))
print(model2.summary())
```

```python
plot_model(model2, show_shapes = True, show_layer_names = True)
```

3rd model with names of the layers

```python
model3 = keras.Sequential(name = 'captain salad')
model3.add(keras.Input(shape=(4,)))
model3.add(layers.Dense(2, activation = 'relu', name = 'tomato'))
model3.add(layers.Dense(3, activation = 'relu', name = 'orange'))
model3.add(layers.Dense(4, name = 'green_tea'))
print(model3.summary())
```

```python
plot_model(model3, show_shapes = True, show_layer_names = True)
```

deleting the last layer

```python
print(f'number of layers :{len(model3.layers)}')
model3.pop()
print(f'number of layers :{len(model3.layers)}')
```

```python
plot_model(model3, show_shapes = True, show_layer_names = True)
```

another way of creating the model using keras

```python
model4 = keras.Sequential(
    [
    keras.Input(shape=(4,)),
    layers.Dense(2, activation = 'relu', name = 'tomato'),
    layers.Dense(3, activation = 'relu', name = 'orange'),
    layers.Dense(4, name = 'green_tea')
])
```

```python
print(model4.summary())
```

```python
plot_model(model4, show_shapes = True, show_layer_names = True)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import plot_model
```

```python
model5 = Sequential()
model5.add(Dense(8, input_dim = 2, activation = 'relu'))
model5.add(Dense(2, activation = 'softmax'))
```

```python
model5.summary()
```

```python
plot_model(model5, show_shapes = True, show_layer_names = True)
```

```python
model6 = keras.Sequential(
    [
    keras.Input(shape=(8,)),
    layers.Dense(12, activation = 'relu', name = 'tomato'),
    layers.Dense(8, activation = 'relu', name = 'orange'),
    layers.Dense(1, name = 'green_tea')
])
```

```python
model6.summary()
```

```python
plot_model(model6, show_shapes = True, show_layer_names = True)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import numpy
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import plot_model
```

```python
seed = 7
numpy.random.seed(seed)
dataset = numpy.loadtxt("/content/pima-indians-diabetes.csv", delimiter = ",")
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
model_final = Sequential()
model_final.add(Dense(12, input_dim = 8, activation = 'relu'))
model_final.add(Dense(8, activation = 'relu'))
model_final.add(Dense(1, activation = 'sigmoid'))
```

```python
model_final.summary()
```
